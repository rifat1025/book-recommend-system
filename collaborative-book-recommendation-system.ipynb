```python
import pandas as pd
import numpy as np
books=pd.read_csv("/kaggle/input/datasets/arashnic/book-recommendation-dataset/Books.csv")
rating=pd.read_csv("/kaggle/input/datasets/arashnic/book-recommendation-dataset/Ratings.csv")
users=pd.read_csv("/kaggle/input/datasets/arashnic/book-recommendation-dataset/Users.csv")
```

```python
books
```

```python
ratings=rating.merge(books, on="ISBN")
```

```python
ratings
```

```python
x=ratings.groupby('User-ID').count()['Book-Rating']>10
literate=x[x].index
```

```python
literate
```

```python
filter_ratings=ratings[ratings['User-ID'].isin(literate)]
```

```python
y=filter_ratings.groupby('Book-Title').count()['Book-Rating']>50
famous_books=y[y].index
```

```python
famous_books
```

```python
final_rating=filter_ratings[filter_ratings['Book-Title'].isin(famous_books)]
```

```python
final_rating.drop_duplicates()
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(pt)
similarity.shape
```

```python
similarity[10]
```

```python
def recommend(book_name):
    
    if book_name not in pt.index:
        print(f"Book '{book_name}' not found in dataset!")
        return
   
    index = int(np.where(pt.index == book_name)[0][0])
    
   
    sim_scores = list(enumerate(similarity[index]))
    
   
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:7]
    
   
    for i in sim_scores:
        print(pt.index[i[0]])
    
  
```

```python
recommend('Think And Grow Rich')
```
